```python
"""
  Week 8 Lab
  Level: Intermediate
  Name: Sri Thummalapalli
  UID: u1514823
"""

import os

EXPENSES_FILE = "expenses.csv"

def display_menu():
  print("\nExpense Tracker Menu:")
  print("1. Add Expense")
  print("2. View Expenses")
  print("3. Edit Expense")
  print("4. Delete Expense")
  print("5. Exit")
  return input("Enter your choice (1/2/3/4/5):")

def add_expense():
  date = input("Enter the date (YYYY-MM-DD):")
  category = input("Enter the category:")
  description = input("Enter the description:")
  amount = input("Enter the amount:")

  #Validate the amount is numeric
  try:
    amount = float(amount)
  except ValueError:
    print("Invalid amount. Please enter a numeric value.")
    return

  with open(EXPENSES_FILE, "a") as file:
    file.write(f"{date},{category},{description},{amount}\n")
  print("Expense added successfully!")

#Function to view all expenses
def view_expenses():
  if not os.path.exists(EXPENSES_FILE):
    print("No expenses found.")
    return

    with open(EXPENSES_FILE, "r") as file:
      expenses  = file.readlines()

    if not expenses:
      print("No expenses found.")
      return

    print("\nRecorded Expenses:")
    for idx, expense in enumerate(expenses,1):
      date, category, description, amount = expense.strip().split(",")
      print(f"{idx}. Date: {date}, Category: {category}, Description: {description}, Amount: ${amount}")

#Function to edit an expense
def edit_expense():
  view_expenses()

  try:
    expense_choice = int(input("Enter the expense number to edit:"))
  except ValueError:
    print("Invalid input. Please enter a valid expense number.")
    return

  with open(EXPENSES_FILE, "r") as file:
    expenses = file.readlines()

  if expense_choice < 1 or expense_choice > len(expenses):
    print("Invalid expense number.")
    return

  expense = expenses[expense_choice - 1].strip().split(",")
  print("\nCurrent Expense Details:")
  print(f"Date: {expense[0]}, Category: {expense[1]}, Description: {expense[2]}, Amount: ${expense[3]}")

  date = input(f"Enter the new date (YYYY-MM-DD) (press enter to keep '{expense[0]}'): " ) or expense[0]
  category = input(f"Enter the new category (press enter to keep '{expense[1]}'):") or expense[1]
  description = input(f"Enter the new description (press enter to keep '{expense[2]}'):") or expense[2]
  amount = input(f"Enter the new amount (press enter to keep '{expense[3]}'):") or expense[3]

  expenses[expense_choice - 1] = f"{date},{category},{description},{amount}\n"

  with open(EXPENSES_FILE, "w") as file:
    file.writelines(expenses)
  print("Expense edited successfully!")

#Function to delete and expense
def delete_expense():
  view_expenses()

  try:
    expense_choice = int(input("Enter the expense number to delete:"))
  except ValueError:
    print("Invalid input. Please enter a valid expense number.")
    return

  with open(EXPENSES_FILE, "r") as file:
    expenses = file.readlines()

  if expense_choice < 1 or expense_choice > len(expenses):
    print("Invalid expense number.")
    return

  del expenses[expense_choice - 1]

  with open(EXPENSES_FILE, "w") as file:
    file.writelines(expenses)
  print("Expense deleted successfully!")

def expenses_tracker():
  while True:
    display_menu()
    try:
      choice = int(input("Enter your choice (1/2/3/4/5):"))
    except ValueError:
      print("Invalid input. Please enter a number.")
      continue

    if choice == 1:
      add_expense()
    elif choice == 2:
      view_expenses()
    elif choice == 3:
      edit_expense()
    elif choice == 4:
      delete_expense()
    elif choice == 5:
      print("Exiting the Expense Tracker. Goodbye!")
      break
    else:
      print("Invalid choice. Please select a valid option (1-5).")

if __name__ == "__main__":
  expenses_tracker()
```
